```python
class computer:
    type = 'computer'  # class/static variable places in class namespace
    def __init__(self,ram,name): # self is used for object reference, init called every time a object is created
                    # init mainly used to intialize vars
                    # self required to resolve object reference, for which object it is used
        self.ram = ram #instance variable, different for different object
        self.name = name
    def details(self):
        print(self.ram,self.name,self.type)
    def compare(self,other):
        if self.ram==other.ram and self.name==other.name:
            return True
        else:
            return False
     
c1 = computer(4,'hp') # objects created in heap memory
c2 = computer(8,'dell')
c3 = computer(4,'hp')
computer.details(c1)
computer.details(c1)
c1.details()  # same as Computer.details(c1)
c2.details()
c1.type = 'laptop'  # can change class var for specific object, shouldn't do that 
c1.details()  # same as Computer.details(c1)
c2.details()
computer.type = 'lappy'  # stat var changed, chaged for c2 not for c1 bcoz c1 overrided the default one
c1.details()  # same as Computer.details(c1)
c2.details()
print(c1.compare(c2),"value of c1 and c2 are different")
print(c1.compare(c3), "values of c1 and c3 are same")
```

```python
# type of methods in class

class student:
    school ='Spring Dale'
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    def avg(self):   #instace method
        return (self.m1+self.m2)/2
    def get_m1(self): # accessor instance method
        return self.m1
    def set_m1(self,value): # mutators instance method
        self.m1 = value
    @classmethod   # decorator for letting pythin know that it is class method
    def schoolname(cls):  # take class as input, "cls", mainly used for class vars
        print(f"school name is {cls.school}")
    @staticmethod   # decorator for letting pythin know that it is static method
    def classinfo(): # no input, only for informative purpose,doesn't use any var
        print("This is a student class. Created for learning purpose")

s1 = student(34,99)
s2 = student(56,68)
student.classinfo() 
student.schoolname()
s1.get_m1()
print(s1.avg())
s1.set_m1(89)
print(s1.avg())
```

```python
# Inheritence

class A:
    def __init__(self):
        print("Class A Init")
    def f1(self):
        print("Class A f1")
    def f2(self):
        print("class A f2")
class B(A):  # B inherits A
    def __init__(self):
        print("Class B Init")
    def f1(self):
        print("Class B f1")
    def f3(self):
        print("class B f3")
class C(B):  # C inherits B , Multilevel Inheritence
    def __init__(self):
        super().__init__()
        print("Class C Init")
    def f1(self):
        print("Class C f1")
    def f4(self):
        print("class C f4")
class D:  
    def __init__(self):
        print("Class D Init")
    def f1(self):
        print("Class D f1")
    def f5(self):
        print("class D f5")
class E(A,D): # Multiple inheritence
    def __init__(self):
        #super().__init__()
        print("Class E Init")
    def f1(self):
        super().f1()
        print("Class E f1")
    def f6(self):
        print("class E f6")

c = C()  # super will call parent, if there is no __init__ in C then by default B then A..if not found in B
e = E()  # MRO - method resolution order... left to right... first is A then D..so go in A if not found then D
e.f1()
```

```python
# Duck Typing
class PyCharm:
    def execute(self):
        print("PyCharm")
class Spyder:
    def execute(self):
        print("Spyder")
class laptop:
    def code(self,ide):
        ide.execute()
ide = PyCharm()
l = laptop()
l.code(ide)
ide1 = Spyder()
l.code(ide1)
```

```python
# operator overloading
class student:
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    def __add__(self,other):
        return student(self.m1+other.m1,self.m2+other.m2)
    def __gt__(self,other):
        if (self.m1+self.m2) > (other.m1+other.m2):
            return True
        else:
            return False
    def __str__(self):  # for printing the object , should return a string, if not overrided, print address by default
        return str(self.m1) + ' ' +str(self.m2)
s1 = student(50,70)
s2 = student(40,70)
s3 = s1 + s2
print(s3.m1,s3.m2)
if s1 > s2:
    print("s1 wins")
else:
    print("s2 wins")
print(s2)
```

```python
# Abstract class concept
# Atleast one abstract method
# Object can't be created for abstract class
from abc import ABC,abstractmethod # ABC... abstract base class
class animal(ABC):
    @abstractmethod
    def legs(self,x):
        pass
    def info(self):
        print("This class implements abstract method, but this is not abstract method")
class dog(animal):
    def legs(self,x):
        print(f"Dog has {x} legs")
class cat(animal):
    def legs(self,x):
        print(f"Cat has {x} legs")

#a = animal() # error as we can't instaniate abstract class
d = dog()
d.legs(5)
d.info() 
c = cat()
c.legs(4)
c.info() 
```

```python
# Generators
# we need to create __iter__ and __next__manually
# generators will give you iterators

def topten():
    yield 1
    yield 2
    n = 1
    while n<=10:
        yield n*n
        n+=1
values = topten()
print(next(values)) # print 1
print(values.__next__()) # print 2
for i in values:   # print 1-10 squaresx
    print(i)
```

```python
# Multithreading
from threading import *
from time import sleep
class hello(Thread):
    def run(self):
        for i in range(5):
            print("hello")
            sleep(1)
class hi(Thread):
    def run(self):
        for i in range(5):
            print("hi")
            sleep(1)
t1 = hello()
sleep(0.2) # so that both thread doesn't reach to cpu at same time
t2 = hi()
t1.start() # start thread t1.....by default it will call run method of t1
t2.start() # start thread t2.....by default it will call run method of t2
t1.join() # wait for t1 to complete
t2.join() # eait for t2 to complete
print("Bye") # runs in main thread
```
